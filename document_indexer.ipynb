```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
#Importera data

data_raw = pd.read_pickle('ICAAD.pkl')
data = data_raw[data_raw['Sexual.Assault.Manual'] | data_raw['Domestic.Violence.Manual']]

```

```python
her = r' ( she | her |complainant|victim|girl|woman|complainant)+'
he = r'(alleged|accused| he | him | his )'
NOT = r'(neither|didn\'t|not|no|deni)'
defend =r' (shout|defend|fight|resist)'
credible = r' (credible|reliable|truthful|honest)'
exaggerate = r' (exaggerate|falsify|extrort|fabricate)'
suffers = r' (suffers|illness|shock)'
provider = r'(bread winner)' 
safe = r'(safe|secur|comfort)'

un = r'(un|in|a|di|ill|non|dis)'
uncredible = un+credible

s = r'[\w ]*'

bad = {
r'child' +s+ NOT +s+ safe: -1,
her +s+ NOT +s+ defend: 1,
her +s+ NOT +s+ credible: 1,
her +s+ credible: -1,
her +s+ uncredible: 1,
her +s+ exaggerate: 1,
r'provoce': 1,
r'virgin': 1,
r'defilement': 0.1,
her + r'[\w ]+free consent': 1,
NOT +s+ her + r'[\w ]+free consent': -2,
r'disbelieve[\w ]+'+her: 1,
NOT +s+ r'disbelieve[\w ]+'+her: -2,
r'the complainant': 0.1,
NOT +s+ r'doubt' +s+ her: -2,
r' doubt' +s+ her: 1,
r' lie': 1,
her +s+ NOT +s+ r' lie': -1,
r' affair': 1,
r'illicit affair': 1,
r'to light earlier': 1,
her +s+ 'story': 1,
r'story' +s+ her: 1,
her + r'wanted to conceal': 1,
r'allegation of force': 1,
r'place little weight[\w ]+'+her: 1, 
r'believe[\w ]+'+her+r'?': 1,
NOT +s+ r'proved beyond reasonable doubt': 2,
he +s+ suffers: 1,
r'village headman': 1,
he +s+ provider: 1
}
r'child' +s+ NOT +s+ safe
```

```python
def print_match(case):
    print('case', case)
    print('####################################################')
    points = 0
    for frase in bad:
        m = re.findall(frase, data.loc[case,'text'])
        if len(m) == 0:
            pass
            #print(None)
        else:
            print(frase)
            for found in m:
                points+=bad[frase]
                print(found)
            print(' Found: '+ str(len(m)) +' times.')
    print(points)
    print('####################################################')


point = {}
for i, row in data.iterrows():
    #print('case', i)
    points = 0

    for frase in bad:
        m = re.findall(frase, row['text'])
        if m is None:
            pass
            #print(None)
        else:
            for found in m:
                points+=bad[frase]
            #print(frase, m.group(0), ' Found: '+ len(m) +' times.')
    point[i] = points/len(row['text'])
    #print(points)
    #print('####################################################')
        
```

```python
for p in sorted(point, key=lambda x: point[x]):
    print(p, point[p])
```
